```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

### import restaurant inspection data

```python
rid = pd.read_csv("../data/NYCRIR.csv")
```

```python
list(rid.columns.values)
```

```python
rid.BORO.unique()
```

### Focus on Bronx Data

```python
BxRid = rid[rid['BORO']=="BRONX"] 
```

```python
rid.GRADE.unique()
```

```python
BxRid = BxRid[BxRid['GRADE']!="Not Yet Graded"] ## Remove stores that have not been graded yet
```

```python
BxRid = BxRid[pd.notnull(BxRid["GRADE"])] ## Remove Stores that have no grade
```

```python
rid.SCORE.unique()
```

```python
BxRid=BxRid[pd.notnull(BxRid["SCORE"])] ## Remove stores with no score
```

```python
BxRid["SCORE"].describe()
```

```python
BxRid["GRADE"] = BxRid["GRADE"].astype("category",categories = ["A","B","C","P","Z"], ordered = True) ## redefine score levels
```

```python
BxRid = BxRid.reset_index(drop=True)
BxRid.head(10)
```

### Grades 

```python
BxRid["GRADE"].value_counts().plot(kind = "pie")
```

```python
import seaborn as sns
sns.set(style="whitegrid", color_codes=True)
```

```python
sns.stripplot(x="GRADE", y = "SCORE", data = BxRid, jitter = True)
```

```python
sns.stripplot(x="GRADE", y = "SCORE", hue = "CRITICAL FLAG", data = BxRid, jitter = True)
```
